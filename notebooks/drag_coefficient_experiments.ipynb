```python
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import netCDF4 as nc
import locale

%matplotlib inline
np.warnings.filterwarnings('ignore')
locale.setlocale(locale.LC_TIME, "es_ES")

# ---
bomm = "bomm1_per1"
number_of_minutes = 30 # 30 min or 10 min
#
basepath = "/Volumes/Boyas/bomm_database/data"
filename = f"{basepath}/{bomm}/level2/{bomm}_level2_{number_of_minutes}min.nc"
dataset = nc.Dataset(filename)
```

```python
U10N = dataset["U10N"][:]
ustar = dataset["ustar"][:]
rhoa = dataset["rhoa"][:]
taux = - rhoa * dataset["uw"][:]
tauy = - rhoa * dataset["vw"][:]
```

```python
fig, (ax1, ax2) = plt.subplots(2, dpi=150, sharex=True, sharey=True)
ax1.plot(U10N, taux, ".", color="red", ms=1)
ax2.plot(U10N, tauy, ".", color="green", ms=1)
#
ax1.axhline(0, color="k")
ax2.axhline(0, color="k")
#
ax1.set_ylabel("$\\tau_{x}\;\mathrm{[N/m^2]}$")
ax2.set_ylabel("$\\tau_{y}\;\mathrm{[N/m^2]}$")
ax2.set_xlabel("$U_{10N}\;\mathrm{[m/s]}$")
```

```python
angle = np.arctan2(tauy, taux) * 180/np.pi
fig, ax = plt.subplots(1, dpi=150)
ax.plot(U10N, angle, ".", ms=1)
ax.axhline(0, color="k")
ax.set_ylim((-180,180))
```
